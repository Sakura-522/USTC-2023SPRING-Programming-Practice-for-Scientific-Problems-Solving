```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 100, 0.01)             # 生成x的取值
x1 = np.arange(1, 10000, 1000).tolist()             # 生成x的取值
g, h, k = np.log2(x), 1.05**x, 100/x    # g(x),h(x),k(x) 的值

plt.rcParams.update({"font.size": 12, "axes.labelsize": 14}) # 设置字体和数字大小

plt.figure(dpi=200) # 生成图片的DPI
plt.plot(x, x, color='r', label='f(x)=x', markevery=x1 , marker='o',markersize=5)            # f(x)=x
plt.plot(x, g, color='c', label='g(x)=log(x)',
         markevery=x1 , marker='s', linestyle='--', markersize=5)                             # g(x)=log(x)
plt.plot(x, h, color='b', label='h(x)=1.05^x',
         markevery=x1 , marker='*', linestyle=':', markersize=5)                              # h(x)=1.05^x
plt.plot(x, k, color='orange', label='k(x)=100/x',
         markevery=x1 , marker='D', linestyle='-.', markersize=5)                             # k(x)=100/x


# 标注曲线
plt.annotate('f(x)=x', xy=(40, 40), xytext=(35, 60),
             xycoords='data',
             arrowprops=dict(facecolor='red', shrink=0.05)
             )

plt.annotate('g(x)=log(x)', xy=(90, 7), xytext=(85, 25),
             xycoords='data',
             arrowprops=dict(facecolor='c', shrink=0.05)
             )

plt.annotate('h(x)=1.05^x', xy=(60, 20), xytext=(55, 40),
             xycoords='data',
             arrowprops=dict(facecolor='b', shrink=0.05)
             )

plt.annotate('k(x)=100/x', xy=(3, 60), xytext=(10, 70),
             xycoords='data',
             arrowprops=dict(facecolor='orange', shrink=0.05)
             )


x_scale = plt.MultipleLocator(10)
y_scale = plt.MultipleLocator(20)
ax = plt.gca()
ax.xaxis.set_major_locator(x_scale)  # 设置x轴刻度大小
ax.yaxis.set_major_locator(y_scale)  # 设置y轴刻度大小
plt.xlim((0, 100))                   # 设置x轴刻度范围
plt.ylim((0, 140))                   # 设置y轴刻度范围
plt.xticks(fontsize=10)              # 设置x轴刻度字体大小
plt.yticks(fontsize=10)              # 设置y轴刻度字体大小

ax.set_xlabel(..., fontsize=15)      # 设置xlabel的字体大小
ax.set_ylabel(..., fontsize=15)      # 设置ylabel的字体大小
plt.xlabel('x', loc='right')         # 设置xlabel
plt.ylabel('y', loc='top')           # 设置ylabel

plt.grid(axis='both', alpha=0.5)     # 显示网格线
plt.title('Function Image', loc='center', fontsize=13)  # 图片标题设置

# 图例设置
plt.legend(bbox_to_anchor=(0, 1.06, 1, 0.2), loc='lower left',
           mode='expand', framealpha=0.5, ncol=2)

# 显示图片
plt.show()
```
